```python
import pandas as pd
main = pd.read_csv('/Users/hazel/Documents/credit_project_data/Jen-Chao Huang/main_x_label_0727.csv')
prev = pd.read_csv('/Users/hazel/Documents/credit_project_data/Jen-Chao Huang/prev_x_label_0727.csv')
```

```python
main.drop(['Loan_count','late_count_inLoan','INSTALMENT_count','late_count_total'], axis=1, inplace=True)
```

```python
main
```

```python
prev
```

```python
main_prev = prev.merge(main,left_on='SK_ID_CURR', right_on='SK_ID_CURR', how='left')
main_prev
```

```python
main_prev['late_count_inLoan / Loan_count'] = main_prev['late_count_inLoan']/main_prev['Loan_count']
main_prev
```

```python
main_prev['late_count_inLoan / Loan_count'] = pd.DataFrame(main_prev['late_count_inLoan / Loan_count'].apply(lambda x:1 if x>0 else x))
main_prev
```

```python
main_prev.drop(['Loan_count','late_count_inLoan','INSTALMENT_count','late_count_total'],axis=1,inplace=True)
#main_prev.to_csv('/Users/hazel/Documents/credit_project_data/Jen-Chao Huang/main_prev.csv')
```

```python
from numpy import mean
from numpy import std
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
X = main_prev.iloc[:,:-1].drop(['PREV_CRE/APP_max','SK_ID_CURR'],axis=1)
y = main_prev[['late_count_inLoan / Loan_count']]
# create pipeline
rfe = RFE(estimator=RandomForestClassifier(n_jobs=-1), n_features_to_select=10)
model = RandomForestClassifier()
pipeline = Pipeline(steps=[('s',rfe),('m',model)])
pipeline.fit(X,y)
# evaluate model
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
n_scores = cross_val_score(pipeline, X, y, scoring='accuracy', cv=3, n_jobs=-1, error_score='raise')
# report performance
print('Accuracy: %.3f (%.3f)' % (mean(n_scores), std(n_scores)))
```

```python
from numpy import mean
from numpy import std
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
X = main_prev.iloc[:,:-1].drop(['PREV_CRE/APP_max','SK_ID_CURR'],axis=1)
y = main_prev[['late_count_inLoan / Loan_count']]
# create pipeline
estimator = LogisticRegression()
rfe = RFE(estimator, n_features_to_select=20)
model = RandomForestClassifier()
pipeline = Pipeline(steps=[('s',rfe),('m',model)])
pipeline.fit(X,y)
# evaluate model
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
n_scores = cross_val_score(pipeline, X, y, scoring='accuracy', cv=3, n_jobs=-1, error_score='raise')
# report performance
print('Accuracy: %.3f (%.3f)' % (mean(n_scores), std(n_scores)))
```
