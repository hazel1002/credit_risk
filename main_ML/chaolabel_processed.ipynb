# 創造late_count_inLoan / Loan_count標籤

```python
import pandas as pd
chao_label = pd.read_csv('/Users/hazel/Documents/credit_project_data/Jen-Chao Huang/processed_allHC_label.csv')
chao_label
```

```python
chao_label['late_count_inLoan / Loan_count'] = chao_label['late_count_inLoan']/chao_label['Loan_count']
chao_label
```

```python
chao_label['late_count_inLoan / Loan_count'] = pd.DataFrame(chao_label['late_count_inLoan / Loan_count'].apply(lambda x:1 if x>0 else x))
chao_label
```

```python
chao_label.drop(['Loan_count','late_count_inLoan','INSTALMENT_count','late_count_total'],axis=1,inplace=True)
chao_label
```

```python
chao_label.drop(chao_label.filter(regex="Unname"),axis=1, inplace=True)
chao_label.to_csv('/Users/hazel/Documents/credit_project_data/Jen-Chao Huang/late_count_inLoan_Loan_count.csv',index=False)
```

# 創造late_count_total / INSTALMENT_count標籤 (三元分類)

```python
import pandas as pd
chao_label = pd.read_csv('/Users/hazel/Documents/credit_project_data/Jen-Chao Huang/processed_allHC_label.csv')
chao_label
```

```python
chao_label['late_count_total / INSTALMENT_count'] = chao_label['late_count_total']/chao_label['INSTALMENT_count']
chao_label
```

```python
chao_label['late_count_total / INSTALMENT_count'].describe()
```

```python
chao_label['late_count_total / INSTALMENT_count'] = pd.DataFrame(chao_label['late_count_total / INSTALMENT_count'].apply(lambda x:2 if x>0.077670 else x))
chao_label['late_count_total / INSTALMENT_count'] = pd.DataFrame(chao_label['late_count_total / INSTALMENT_count'].apply(lambda x:1 if x<0.077670 and x>0 else x))
chao_label['late_count_total / INSTALMENT_count'].hist()
```
