```python
import pandas as pd
prev = pd.read_csv('/Users/hazel/Documents/credit_project_data/Jen-Chao Huang/prev_x_label_0727.csv')
prev
```

```python
prev['late_count_inLoan / Loan_count'] = prev['late_count_inLoan']/prev['Loan_count']
prev
```

```python
prev['late_count_inLoan / Loan_count'] = pd.DataFrame(prev['late_count_inLoan / Loan_count'].apply(lambda x:1 if x>0 else x))
prev
```

```python
prev.drop(['Loan_count','late_count_inLoan','INSTALMENT_count','late_count_total'],axis=1,inplace=True)
prev
```

```python
features = prev[['PREV_ANNUITY_median', 'PREV_DOWN_PAYMENT_median',
       'DAYS_DECISION_median', 'CNT_PAYMENT_median', 'DAYS_PERIOD_median',
       'PAYMENT_CASH_BANK', 'YIELD_low', 'HOUR_APPR_PROCESS_START',
       'WEEKDAY_APPR_PROCESS_START', 'NAME_TYPE_SUITE']]
features
```

```python
import numpy as np
from sklearn.linear_model  import LogisticRegression
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.metrics import f1_score
from sklearn.model_selection import cross_val_score

X = prev.iloc[:,:-1].drop(['PREV_CRE/APP_max'],axis=1)
y = prev[['late_count_inLoan / Loan_count']]

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.5) #random_state 種子值

scaler = preprocessing.StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)

model=LogisticRegression(C=1.6)
model.fit(X_train,y_train) #就是在調參數

X_test_nor = scaler.transform(X_test)
y_pred = model.predict(X_test_nor)

accuracy = accuracy_score(y_test, y_pred)
num_correct_samples = accuracy_score(y_test, y_pred, normalize=False)
con_matrix = confusion_matrix(y_test, y_pred)
f1_score = f1_score(y_test, y_pred, average=None)
cross_val_score = cross_val_score(model,X,y,cv=3,scoring='accuracy')

print('number of correct sample: {}'.format(num_correct_samples))
print('accuracy: {}'.format(accuracy))
print('con_matrix: {}'.format(con_matrix))
print('f1_score: {}'.format(f1_score))
print('cross_val_score: {}'.format(cross_val_score))
print('cross_val_score_mean: {}'.format(cross_val_score.mean()))
print('訓練集：{}'.format(model.score(X_train, y_train)))
print('測試集：{}'.format(model.score(X_test,y_test)))
```
