```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/Users/hazel/Documents/credit_project2/bureau_balance.csv')
```

```python
df
```

```python
m_max = df.groupby(['SK_ID_BUREAU'])['MONTHS_BALANCE'].agg({})
```

```python
m_max = pd.DataFrame(m_max)
m_max
```

```python
df2 = df[df['STATUS'] == 'C']
```

```python
df2
```

```python
s_min = df2.groupby(['SK_ID_BUREAU'])['MONTHS_BALANCE'].agg(min)
```

```python
s_min = pd.DataFrame(s_min)
s_min
```

```python
# 把有繳清的人merge
merge = pd.merge(m_max,s_min,how='outer',left_index=True,right_index=True)
merge['Start_to_End'] = merge['MONTHS_BALANCE_x'] - merge['MONTHS_BALANCE_y']
```

```python
merge
```

```python
merge['CLOSE'] = merge['Start_to_End'].isnull()
merge
```

```python
scale_mapper = {False:1, 
                True:0}
merge['CLOSE'] = merge['CLOSE'].replace(scale_mapper)
merge
```

```python
merge.corr()
```

```python
merge.drop(['MONTHS_BALANCE_y'],axis=1,inplace=True)
merge
```

```python
merge.to_csv('bureau_balance_clear.csv')
```
