##開始合併表格製作分數

##製作簡易label

```python
filter=df_mix["SK_ID_CURR"][(df_mix['bureau_CREDIT_DAY_OVERDUE']>0) | (df_mix['PC_SK_DPDMix_Max']>0) | (df_mix['NAN']>0)]
```

```python
filter=filter.to_frame()
```

```python
filter
```

```python
len(filter) 
```

```python
lists=[i*0+1 for i in range(len(filter))]
```

```python
lists
```

```python
filter['simple_labelx']=lists
```

```python
filter
```

```python
df_mix_s=pd.merge(df_mix,filter,how='outer')
```

```python
df_mix_s['simple_labelx']=df_mix_s['simple_labelx'].fillna(0)
```

```python
df_mix_s['simple_labelx'].value_counts()
```

##製作分數label

```python
list=[i*0+100 for i in range(len(df_mix_s))]
```

```python
len(list)
```

```python
df_mix_s['scores']=list
```

```python
df_mix_s['bureau_CREDIT_ACTIVE'][3]>5 and df_mix_s['bureau_CREDIT_ACTIVE'][5]<10
```

```python
for i in range(len(df_mix_s['scores'])):
  if df_mix_s['bureau_CREDIT_ACTIVE'][i]>=5 and df_mix_s['bureau_CREDIT_ACTIVE'][i]<10:
    df_mix_s['scores'][i]-=4
  if df_mix_s['bureau_CREDIT_ACTIVE'][i]>10:
    df_mix_s['scores'][i]-=10
  if df_mix_s['bureau_CREDIT_DAY_OVERDUE'][i]>0:
    df_mix_s['scores'][i]-=10
  if df_mix_s['bureau_CNT_CREDIT_PROLONG'][i]>0:
    df_mix_s['scores'][i]-=10
  if df_mix_s['PC_count'][i]>=5 and df_mix_s['PC_count'][i]<10:
    df_mix_s['scores'][i]-=4
  if df_mix_s['PC_count'][i]>10:
    df_mix_s['scores'][i]-=10
  if df_mix_s['PC_SK_DPDMix_Max'][i]>0:
    df_mix_s['scores'][i]-=10
  if df_mix_s['SK_NUM_PREV'][i]>=5 and df_mix_s['SK_NUM_PREV'][i]<10:
    df_mix_s['scores'][i]-=4
  if df_mix_s['SK_NUM_PREV'][i]>10:
    df_mix_s['scores'][i]-=10
  if df_mix_s['DAYS_LATE_PAYMENT'][i]>=15 and df_mix_s['DAYS_LATE_PAYMENT'][i]<30:
    df_mix_s['scores'][i]-=4
  if df_mix_s['DAYS_LATE_PAYMENT'][i]>30:
    df_mix_s['scores'][i]-=4
  if df_mix_s['AMT_ARREARS'][i]>=1 and df_mix_s['AMT_ARREARS'][i]<5:
    df_mix_s['scores'][i]-=4
  if df_mix_s['AMT_ARREARS'][i]>5:
    df_mix_s['scores'][i]-=10
  if df_mix_s['NAN'][i]>0:
    df_mix_s['scores'][i]-=10
```
